```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random 100000x1 arrays from exponential, uniform, and normal distributions
exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.uniform(low=0.0, high=1.0, size=100000)
normal_data = np.random.normal(loc=0.0, scale=1.0, size=100000)

# Create histograms
plt.hist(exp_data, density=True, bins=100, histtype='step', color='red', label='Exponential')
plt.hist(uniform_data, density=True, bins=100, histtype='step', color='blue', label='Uniform')
plt.hist(normal_data, density=True, bins=100, histtype='step', color='green', label='Normal')

# Adjust plot view
plt.axis([0, 5, 0, 1.5])

# Add legend, title, and labels
plt.legend(loc='upper right')
plt.title('Random Distributions')
plt.xlabel('Value')
plt.ylabel('Density')

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of points
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plot of Z = X^2 + Y^2')

# Show plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Select the specified features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df_selected = df[features]

# Calculate Pearson correlation
pearson_corr = df_selected.corr(method='pearson')

# Calculate Spearman correlation
spearman_corr = df_selected.corr(method='spearman')

# Create heatmaps
plt.figure(figsize=(12, 5))

# Pearson correlation heatmap
plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', cbar=True, square=True, fmt='.2f')
plt.title('Pearson Correlation')

# Spearman correlation heatmap
plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', cbar=True, square=True, fmt='.2f')
plt.title('Spearman Correlation')

plt.tight_layout()
plt.show()
```
